```python
#Create a new data frame with all the values necesary for predicting revenue. 
new_df = pd.concat([df['Text'], df['Description'],df["year "]], axis=1)
new_df["Text"] = new_df['Text'].astype(str)
new_df["Description"] = new_df['Description'].astype(str)
new_df["year "] = new_df['year '].astype(int)
print(new_df.keys())



```

```python
#Cleaning 1: Extracting movie rating. 

new_df['Description'] = new_df['Description'].str.strip()

# define a regular expression pattern to extract the rating
rating_pattern = r'MPAA\s+Rating:\s*(.*?)\s*\('

# extract the rating using the regular expression
new_df['Rating'] = new_df['Description'].apply(lambda x: re.findall(rating_pattern, x))

# clean up the rating text by replacing non-alphanumeric characters with spaces
new_df['Rating'] = new_df['Rating'].apply(lambda x: re.sub(r'\W+', ' ', x[0]) if len(x) > 0 and x[0] != '' else None)

```

```python
#Cleaning 2: Extracting Keywords

# define a regular expression pattern to extract the Keywords
rating_pattern = r'Creative\s+Type:\s*(.*?)\n'

# extract the Keywords using the regular expression
new_df['CreativeType'] = new_df['Description'].apply(lambda x: re.findall(rating_pattern, x, re.DOTALL))

# clean up the Keywords text by replacing non-alphanumeric characters with spaces
new_df['CreativeType'] = new_df['CreativeType'].apply(lambda x: re.sub(r'\W+', ' ', x[0]).strip() if len(x) > 0 and x[0] != '' else None)
```

```python
#Cleaning 3: Extracting Sources 
# define a regular expression pattern to extract the Source
rating_pattern = r'Source:\s*(.*?)\n'

# extract the Source using the regular expression
new_df['Source'] = new_df['Description'].apply(lambda x: re.findall(rating_pattern, x, re.DOTALL))

# clean up the Source text by replacing non-alphanumeric characters with spaces
new_df['Source'] = new_df['Source'].apply(lambda x: re.sub(r'\W+', ' ', x[0]).strip() if len(x) > 0 and x[0] != '' else None)
```

```python
#Cleaning 4: Exctacting Movie Genre
# define a regular expression pattern to extract the Genre
rating_pattern = r'Genre:\s*(.*?)\n'

# extract the Genre using the regular expression
new_df['Genre'] = new_df['Description'].apply(lambda x: re.findall(rating_pattern, x, re.DOTALL))

# clean up the Genre text by replacing non-alphanumeric characters with spaces
new_df['Genre'] = new_df['Genre'].apply(lambda x: re.sub(r'\W+', ' ', x[0]).strip() if len(x) > 0 and x[0] != '' else None)
```

```python
#Cleaning 5: Exctrating Langueges
# define a regular expression pattern to extract the Languages
rating_pattern = r'Languages:\s*(.*?)\n'

# extract the Languages using the regular expression
new_df['Languages'] = new_df['Description'].apply(lambda x: re.findall(rating_pattern, x, re.DOTALL))

# clean up the Languages text by replacing non-alphanumeric characters with spaces
new_df['Languages'] = new_df['Languages'].apply(lambda x: re.sub(r'\W+', ' ', x[0]).strip() if len(x) > 0 and x[0] != '' else None)
```

```python
#Cleaning 6: Extracting Method 
# define a regular expression pattern to extract the Production type
rating_pattern = r'Production\s+Method:\s*(.*?)\n'

# extract the Production type using the regular expression
new_df['Production_type'] = new_df['Description'].apply(lambda x: re.findall(rating_pattern, x, re.DOTALL))

# clean up the Production type text by replacing non-alphanumeric characters with spaces
new_df['Production_type'] = new_df['Production_type'].apply(lambda x: re.sub(r'\W+', ' ', x[0]).strip() if len(x) > 0 and x[0] != '' else None)
```

```python
#Cleaning 7: Running Time 
# define a regular expression pattern to extract the Running Time
rating_pattern = r'Running\s+Time:\s*(.*?)\n'

# extract the Running Time using the regular expression
new_df['Running_Time'] = new_df['Description'].apply(lambda x: re.findall(rating_pattern, x, re.DOTALL))

# clean up the Running Time text by replacing non-alphanumeric characters
```

```python
#Cleaning  8: Sypnopsis 
# strip any leading/trailing white space from the 'Description' column
new_df['Description'] = new_df['Description'].str.strip()

# define a regular expression pattern to extract the synopsis
synopsis_pattern = r'Synopsis(.*?)Metrics'

# extract the synopsis using the regular expression
new_df['Synopsis'] = new_df['Description'].apply(lambda x: re.findall(synopsis_pattern, x, flags=re.DOTALL))

# clean up the synopsis text by replacing non-alphanumeric characters with spaces
new_df['Synopsis'] = new_df['Synopsis'].apply(lambda x: re.sub(r'\W+', ' ', x[0]).strip() if len(x) > 0 else None)

```

```python
#Order text
# Replace spaces and new lines in the Description column
new_df['Description'] = new_df['Description'].str.replace(' ', '').str.replace('\n', '')

# Print keys to check the updated dataframe
print(new_df.keys())
```

```python
#Cleaning 9: Filter budget 

new_df['Description'] = new_df['Description'].str.strip()
# define a regular expression pattern to extract the production budget
budget_pattern = r'Production\s+Budget:\$([\d,]+)'


# extract the production budget using the regular expression
new_df['budget'] = new_df['Description'].apply(lambda x: re.findall(budget_pattern, x))

# convert the extracted value to an integer
new_df['budget'] = new_df['budget'].apply(lambda x: int(x[0].replace('$', '').replace(',', '')) if len(x) > 0 else None)

```

```python
#CLeaning 10: Domestic Boxoffice 
new_df['Description'] = new_df['Description'].str.strip()

# define a regular expression pattern to extract the production budget
budget_pattern = r'AllTimeDomesticBoxOffice.*?\$([\d,]+)'

# extract the production budget using the regular expression
new_df['Revenue'] = new_df['Description'].apply(lambda x: re.findall(budget_pattern, x))

# convert the extracted value to an integer
new_df['Revenue'] = new_df['Revenue'].apply(lambda x: int(x[0].replace('$', '').replace(',', '')) if len(x) > 0 and x[0] != '' else None)

```

```python
#Cleaning 11: Revenue International
new_df['Description'] = new_df['Description'].str.strip()
# define a regular expression pattern to extract the production budget
budget_pattern = r'AllTimeWorldwideBoxOffice.*?\$([\d,]+)'

# extract the production budget using the regular expression
new_df['Revenue_int'] = new_df['Description'].apply(lambda x: re.findall(budget_pattern, x))

# convert the extracted value to an integer
new_df['Revenue_int'] = new_df['Revenue_int'].apply(lambda x: int(x[0].replace('$', '').replace(',', '')) if len(x) > 0 and x[0] != '' else None)

```

```python
#Cleaning 12: Keywords


# Define a regular expression pattern to extract the keywords
keyword_pattern = r'Keywords:(.*?)(?=Source:|Genre:|$)'

# Extract the keywords using the regular expression
new_df['Keywords'] = new_df['Description'].apply(lambda x: re.findall(keyword_pattern, x, re.DOTALL)[0].split(',') if re.findall(keyword_pattern, x, re.DOTALL) else [])

# Clean up the extracted keywords by removing leading/trailing spaces
new_df['Keywords'] = new_df['Keywords'].apply(lambda x: [keyword.strip() for keyword in x])


```

```python
import pandas as pd
import re

Movie_Info["Text3"] = Movie_Info["Text3"].astype('str')

# define a regular expression pattern to extract the lead ensemble members
lead_pattern = r'(?<=Lead Ensemble Members\n)[\s\S]*(?=\n\nSupporting Cast)'

# define a function to apply the regular expression to each row of the data frame
def extract_lead_ensemble_members(text):
    lead_ensemble_members = re.findall(lead_pattern, text)
    if len(lead_ensemble_members) > 0:
        lead_ensemble_members = lead_ensemble_members[0].split('\n\n')
        lead_ensemble_members = [re.sub(r'\n', ' /', member) for member in lead_ensemble_members]
        return lead_ensemble_members
    else:
        return None
#Clean Data
def clean_cast(cast_list):
    if cast_list is None:
        return []
    clean_list = []
    for cast_member in cast_list:
        cast_member = cast_member.strip()
        if cast_member:
            clean_list.append(cast_member.replace('\xa0', ':'))
    return clean_list

# apply the function to the 'Description' column and create a new column with the results
Movie_Info['Lead_Ensemble_Members'] = Movie_Info['Text3'].apply(extract_lead_ensemble_members)
Movie_Info["Clean_Lead_Ensemble_Members"] = Movie_Info["Lead_Ensemble_Members"].apply(clean_cast)

```

```python
def split_actors_and_characters(row):
    if row[0] == '':
        return [], []
    actors_and_characters = row[0].split(';')
    lead_actors = actors_and_characters[0].split(' and ') if ' and ' in actors_and_characters[0] else actors_and_characters[0].split(', ')
    lead_characters = [] if len(actors_and_characters) == 1 else [character.strip() for character in actors_and_characters[1].split(',')]
    return lead_actors, lead_characters


# apply function to each row of the dataframe
merged_df[['Lead_Actors', 'Lead_Characters']] = merged_df['merged_column'].apply(lambda row: pd.Series(split_actors_and_characters(row)))

print(merged_df)
```

```python
def extract_lead_actors_characters(row):
    lead_actors = []
    lead_characters = []
    for string in row:
        actors = [actor.strip() for actor in string.split(';') if actor.strip()]
        if actors:
            lead_actors.append(actors[0])
        else:
            lead_actors.append('')
        characters = [character.strip() for character in string.split(';')[1:] if character.strip()]
        lead_characters.append(characters)
    return pd.Series([lead_actors, lead_characters])

merged_df[['Lead_Actors', 'Lead_Characters']] = merged_df['merged_column'].apply(extract_lead_actors_characters)

merged_df[['Supporting_Actors', 'Supporting_Characters']] = merged_df['Clean_Supporting_Cast'].apply(extract_lead_actors_characters)
```

```python
import ast
merged_df['Clean_Production_Technical_Credits2'] = merged_df['Clean_Production_Technical_Credits2'].apply(lambda x: ast.literal_eval(x))
```

```python
for index, row in merged_df.iterrows():
    for key in row['Clean_Production_Technical_Credits2'].keys():
        if key not in merged_df.columns:
            merged_df[key] = pd.Series(dtype=object)
        merged_df.at[index, key] = row['Clean_Production_Technical_Credits2'][key]

print(merged_df)
```

```python
new_df = merged_df.dropna(thresh=700, axis=1)
print(new_df.head())
```

```python
# extract the rating using regular expressions
new_df['Rating_Code'] = new_df['Rating'].str.extract(r'^(G|PG|PG 13|R|NC 17)')

# extract the reason for rating using regular expressions
new_df['Rating_Reason'] = new_df['Rating'].str.extract(r'^((?:G|PG|PG 13|R|NC 17)\s+)(.*)$')[1]

```

```python
# function to merge the lists
def merge_lists(row):
    merged_list = []
    for sublist in row:
        merged_list += sublist
    return merged_list

# apply the function to each row
new_df['Supporting_Characters'] = new_df['Supporting_Characters'].apply(ast.literal_eval)
new_df['Supporting_Characters2'] = new_df['Supporting_Characters'].apply(merge_lists)


new_df['Lead_Characters'] = new_df['Lead_Characters'].apply(ast.literal_eval)
new_df['Lead_Characters2'] = new_df['Lead_Characters'].apply(merge_lists)

# print the result
print(new_df.head())
```

```python
import re
# define a regular expression pattern to extract the release dates
release_pattern = r'InternationalReleases:(.*?)\('

# extract the release dates using the regular expression
new_df['ReleaseDates'] = new_df['Description'].apply(lambda x: re.findall(release_pattern, x, re.DOTALL))

# clean up the release dates text by replacing non-alphanumeric characters with spaces
new_df['ReleaseDates'] = new_df['ReleaseDates'].apply(lambda x: re.sub(r'\W+', ' ', x[0]).strip() if len(x) > 0 and x[0] != '' else None)

print(new_df.head(20))
```
